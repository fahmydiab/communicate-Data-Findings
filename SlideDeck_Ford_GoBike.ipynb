# Investigation Overview
I wanted to take a look at the features contained in the dataset that affects the trips durations. 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline
```

```python
df = pd.read_csv('201902-fordgobike-tripdata.csv')
```

```python
df.describe()
```

```python
df.bike_share_for_all_trip = (df.bike_share_for_all_trip == 'Yes')
df.drop(df[df['member_birth_year'] <=1919].index, inplace = True) 
df['duration_minute']=df['duration_sec']/60
df['start_hourofday'] = pd.DatetimeIndex(df['start_time']).hour
weekday = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
weekdaycat = pd.api.types.CategoricalDtype(ordered=True, categories=weekday)
df['start_dayofweek'] = pd.DatetimeIndex(df['start_time']).day_name()
df['start_dayofweek'] = df['start_dayofweek'].astype(weekdaycat)
```

```python
df['start_time'] = pd.to_datetime(df['start_time'])
df['end_time'] = pd.to_datetime(df['end_time'])

df['user_type'] = df['user_type'].astype('category')
df['member_gender'] = df['member_gender'].astype('category')

df.info(null_counts=True)
```

## Trip Duration destribution
Plotting the distribution of trips according to their durations

The trip duration distributoin is skewed steeply to the right, with mode equals over 100,000 for durations between 0 and 10 mins.And the plot shows a mean of 11 mins.

```python
plt.figure(figsize=[10, 5])
plt.hist(data = df, x = 'duration_minute',bins=[0, 10, 20, 30, 40, 50, 100])
plt.title('Distribution of Trip Durations')
plt.xlabel('Duration (min)')
plt.ylabel('Number of Trips')
plt.axvline(df['duration_minute'].mean(), color='r', linestyle='--')
plt.show()
```

## Distribution of users age
Plotting the distribution of users age

The high percent of users have ages between 25-35 years old

```python
binsize = 1
bins = np.arange(0, df['member_birth_year'].astype(float).max()+binsize, binsize)
minBirthYear = df['member_birth_year'].astype(float).min()
maxBirthYear =df['member_birth_year'].astype(float).max()
plt.figure(figsize=[8, 5])
plt.hist(data = df.dropna(), x = 'member_birth_year', bins = bins)
plt.axis([minBirthYear, maxBirthYear, 0, 12000])
plt.xticks([minBirthYear, minBirthYear+10, minBirthYear+20, minBirthYear+30, minBirthYear+40, minBirthYear+50, minBirthYear+60, maxBirthYear], [(2019-minBirthYear), (2019-minBirthYear-10), (2019-minBirthYear-10), (2019-minBirthYear-20), (2019-minBirthYear-30), (2019-minBirthYear-40), (2019-minBirthYear-50), (2019-maxBirthYear)])
plt.gca().invert_xaxis()
plt.title('Distribution of User Age')
plt.xlabel('Age (years)')
plt.ylabel('Number of Users')
plt.show()
```

# The relationship between age and trip durations
The plot shall view the users age with the corresponding trip durations 

Obviously higher durations are for users with ages near 30 years old

```python
plt.figure(figsize=[14,5])

plt.subplot(1, 2, 1)
plt.scatter((2019 - df['member_birth_year']), df['duration_minute'], alpha = 0.25, marker = '.' )
plt.axis([18, 85, 10, 200])
plt.title('Trip Duration and Age')
plt.xlabel('Age')
plt.ylabel('Duaration(min)')

plt.subplot(1, 2, 2)
bins_y = np.arange(0, 50, 5)
bins_x = np.arange(18, 60, 10)
plt.hist2d((2019 - df['member_birth_year']), df['duration_minute'],
           bins = [bins_x, bins_y])
plt.colorbar(ticks=[0, 10, 20, 50]);
plt.show()
```

## Bivariate relation between duration and gender

There are alot of female using long ride trips than male

```python
plt.figure(figsize = [10, 5])
base_color = sb.color_palette()[2]
sb.boxplot(data = df, x = 'member_gender', y = 'duration_minute', color = base_color)
plt.ylim([0, 50])
plt.title('Trip Duration and Gender')
plt.xlabel('Gender')
plt.ylabel('Duration(min)')
plt.show()
```

## Hourly Usage during Weekdays for Male and Female
There are no different usage patterns for the two genders of riders . 
Many trips concentrated around 7-10am and 17-19 pm on work days when used for work commute
Other than the different usage goals, it may also have to do with the traffic situation around these rush hours in the area.

```python
plt.figure(figsize=[8, 8]);

plt.subplot(2, 1, 1)
male = df.query('member_gender == "Male"')
ct_counts = male.groupby(['start_dayofweek', 'start_hourofday']).size()
ct_counts = ct_counts.reset_index(name='count')
ct_counts = ct_counts.pivot(index='start_dayofweek', columns='start_hourofday', values='count')
sb.heatmap(ct_counts, cmap='rocket_r');
plt.title('Male Users', loc='center');
plt.xlabel('Hour of Day');
plt.ylabel('Day of Week');
plt.tight_layout();

plt.subplot(2, 1, 2)
female = df.query('member_gender == "Female"')
st_counts = female.groupby(['start_dayofweek', 'start_hourofday']).size()
st_counts = st_counts.reset_index(name='count')
st_counts = st_counts.pivot(index='start_dayofweek', columns='start_hourofday', values='count')
sb.heatmap(st_counts, cmap='rocket_r');
plt.title('Female Users', loc='center');
plt.xlabel('Hour of Day');
plt.ylabel('Day of Week');
plt.tight_layout();
plt.suptitle('Hourly Usage during Weekdays for Male and Female', fontsize=14, fontweight='semibold', y=1.015);
```

## User gender and their weekly usage
The plot is showing that the majority of trips are done during the weekdays and the most users gender is male, so we can coclude that users take rides for commuting, as the number of rides decrease in weekends.

```python
plt.figure(figsize=[16, 6]);
base_color = sb.color_palette('colorblind')[3];
sb.set_style('darkgrid');

plt.subplot(1, 2, 1);
sorted_vals = df.dropna().member_gender.value_counts()/df.dropna().shape[0];
plt.pie(sorted_vals, explode=[0, 0.1, 0], labels=sorted_vals.index, startangle=90, counterclock=False, autopct='%1.1f%%', shadow=True);
plt.axis('square');

plt.subplot(1, 2, 2);
ax = sb.countplot(data=df, x='start_dayofweek', hue='member_gender', hue_order=sorted_vals.index);
ax.legend(title='User gender');
plt.xticks(rotation=30);
plt.xlabel('');
plt.ylabel('Count');
plt.suptitle('User Genders and Their Weekly Usage', fontsize=14, fontweight='semibold', y=1.015);

```

## Average Weekday Trip Duration for Customers and Subscribers
Average trip duration for rides made by customers are longer than that for subscribers

```python
plt.figure(figsize=[8, 8]);
ax = sb.pointplot(data=df, x='start_dayofweek', y='duration_minute', hue='member_gender', dodge=0.3, markers='.', linestyles="", hue_order=sorted_vals.index);
ax.legend(title='User Gender');
plt.xlabel('');
plt.ylabel('Average Trip Duration (minutes)');
plt.suptitle('Average Weekday Trip Duration for Gender Types', fontsize=14, fontweight='semibold', y=1.015);
```
